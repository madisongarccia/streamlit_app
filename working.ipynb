```python
import plotly.express as px
import pandas as pd
import numpy as np
import folium
import geopy
from geopy.geocoders import Nominatim
```

```python
data = {
    'City': ['Los Angeles', 'San Diego', 'Santa Barbara', 'Long Beach']
}

df = pd.DataFrame(data)

# Create a folium map centered around Southern California
map_sc = folium.Map(location=[34.052235, -118.243683], zoom_start=7)

# Initialize Nominatim geocoder
geolocator = Nominatim(user_agent="southern_california")

# Add markers for each city
for index, row in df.iterrows():
    location = geolocator.geocode(row['City'] + ", Southern California")
    if location:
        folium.Marker(location=[location.latitude, location.longitude], popup=row['City']).add_to(map_sc)

# Display the map
map_sc.save('southern_california_cities_geocoded.html')
map_sc
```
